## Problems

Input table is `logica_home.Salaries`, with salaries of employees of San Francisco by year.

1. The table has a bug: sometimes name have double spaces. Fix the bug by creating an version of the table, where this bug is fixed and double spaces as well as outter spaces are removed. It’s OK to keep only the columns you need for the following problems.

2. Add to this database a table for predicate:
`RoleInfo(role_name:, num_people:, average_pay:)`
that shows the number of people in the role and average pay for people in the role.  Have it ordered by `role_name`. `total_pay` column should be used for `average_pay`.

3. Add to this database a table corresponding to the following predicate:
`PersonInfo(name:, latest_total_pay:, pay_by_year:, pay_relative_to_role:)`
which for each person stores payment they’ve got in the last year that they worked, a list of `{year:, pay:}` structs with pay for each year and how much they get paid relative to the average for the role. Order it by `latest_total_pay` starting from the largest. `total_pay` column should be used for all of `latest_total_pay`, `pay_by_year`, `pay_relative_to_role`. `pay_relative_to_role` should be computed for the last year they were employed.

4. Write predicate `TopPaid(name:, total_pay:)` with 10 people who earned the most in total over all the years they worked.

5. Write a predicate `PayIncrease(person) = increase` showing how much their pay increased from the first to the last year they worked. This predicate may only use `PersonInfo` and must be non-aggregating, i.e. using aggregating expression for getting the increase.
