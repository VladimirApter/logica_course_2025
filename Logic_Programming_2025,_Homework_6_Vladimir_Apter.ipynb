```python
%%logica SaveData

@Engine("duckdb");
@Ground(AustinCrime);
AustinCrime(unique_key:, clearance_date:, description:, location_description:, primary_type:, year:) :-
  `('austin_crime.json')`(unique_key:, clearance_date:, description:, location_description:, primary_type:, year:);

SaveData() += 1 :- AustinCrime();
```

```python
%%logica InefficientPropetyCrimeOnStreet

@Engine("duckdb");

# Grounding predicate without defining it tells Logica that it should be read from the
# current database.
@Ground(AustinCrime);

GetStreet(s) = result :-
  result == RegexpExtract(s, "^\d* (.*)$"),
  !IsNull(result);

InefficientPropetyCrimeOnStreet(street) distinct :-
  AustinCrime(unique_key: a_key, location_description: a, primary_type: "Auto Theft"),
  AustinCrime(unique_key: b_key, location_description: b, primary_type: "Theft"),
  AustinCrime(unique_key: c_key, location_description: c, primary_type: "Burglary"),
  AustinCrime(unique_key: d_key, location_description: d, primary_type: "Burglary"),
  c_key != d_key,
  street == GetStreet(a),
  street == GetStreet(b),
  street == GetStreet(c),
  street == GetStreet(d);


```

```python
%%logica PropertyCrimeOnStreet

@Engine("duckdb");

# Grounding predicate without defining it tells Logica that it should be read from the
# current database.
@Ground(AustinCrime);

GetStreet(s) = result :-
  result == RegexpExtract(s, "^\d* (.*)$"),
  !IsNull(result);

StreetCrime(street:, primary_type:, unique_key:) :-
    AustinCrime(unique_key:, location_description:, primary_type:),
    street == GetStreet(location_description);

Theft(street) :- StreetCrime(street:, primary_type: "Theft", unique_key:);

AutoTheft(street) :- StreetCrime(street:, primary_type: "Auto Theft", unique_key:);

Burglaries(street:, burglaries_count? Count = unique_key) distinct :-
    StreetCrime(street:, primary_type: "Burglary", unique_key:);

PropertyCrimeOnStreet(street) distinct :-
    Theft(street),
    AutoTheft(street),
    Burglaries(street:, burglaries_count:),
    burglaries_count >= 2;
```

N — число записей в AustinCrime  
M — число уникальных улиц

InefficientPropetyCrimeOnStreet:
- Время: ~O(N^4) - делаем 4 вложенных Join
- Память: ~O(N^4) - храним промежуточные результаты соединений  
- Стоимость: ~O(N^4) - в худшем случае, из за 4-x Join

PropertyCrimeOnStreet:  
- Время: ~O(N) - делаем однократный проход с агрегацией
- Память: ~O(M) - храним только хеш-таблицу уникальных улиц
- Стоимость: ~O(N) - делаем однократный проход с агрегацией
